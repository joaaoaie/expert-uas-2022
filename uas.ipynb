```python
t_data = n_data
a = 1
b = 1
# Inisialisasi variabel
```

```python
attractiveness = np.zeros((6, 6))

counter = 0
for r in range(6):
    for c in range(6):
        attractiveness[r][c] = t_data.iloc[counter]['Jarak']
        counter = counter + 1

pheromones = np.ones((6, 6))

print(attractiveness)
print(pheromones)

# Menampilkan nilai attractiveness serta pheromones
```

```python
def run_aco():
    for i in range(5): #Jumlah iterasi
        # print(f"i = {i}")
        final_path = np.array([[]])
        for j in range(2): #Jumlah semut/ambulan
            # print(f"j = {j}")
            chosen_path = np.array([0])
            for k in range(5):
                # print(f"k = {k}")
                accP = np.array([[],[]])

                for l in range(6):
                    # print(f"l = {l}")
                    isContinue = False
                    for temp in chosen_path: #Ngecek ud pernah dilewatin ato blm path ny
                        if temp == l:
                            isContinue = True
                            break
                    if isContinue == True: #kalo ud dilewatin, iterasi loop di skip
                        # print("Skip")
                        continue
                    p = (pheromones[chosen_path[-1]][l]**a) * (attractiveness[chosen_path[-1]][l]**b)
                    print(f"{chosen_path[-1]} - {l} p: {p}")
                    accP = np.append(accP, np.array([[p],[l]]), axis=1)
                total = 0
                for p in accP[0]:
                    total += p
                
                for n in range(accP[0].size):
                    accP[0][n] = accP[0][n] / total
                
                # print("\nSebelum akumulasi: ", accP[0], "/iteration")

                for n in reversed(range(accP[0].size)):
                    for m in range(n):
                        accP[0][n] = accP[0][n] + accP[0][m]
                    
                r = random.Random().random()

                # print("\nSesudah akumulasi: ", r, " - ", accP, "\n")

                for n in range(accP[0].size):
                    if r <= accP[0][n]:
                        chosen_path = np.append(chosen_path, [int(accP[1][n])])
                        print("Chosen Path: ", chosen_path)
                        break
                print()
        final_path = np.append(final_path, [chosen_path])
        return final_path.astype(int)

print(run_aco())
    
```
