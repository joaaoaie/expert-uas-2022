```python
t_data = n_data.copy()
a = 1
b = 1
# Inisialisasi variabel
```

```python
attractiveness = np.zeros((6, 6))
waktu = np.zeros((6, 6))
covid = np.zeros((6, 6))

counter = 0
for r in range(6):
    for c in range(6):
        attractiveness[r][c] = t_data.iloc[counter]['Jarak']
        waktu[r][c] = t_data.iloc[counter]['Waktu']
        covid[r][c] = t_data.iloc[counter]['KonfirmasiCovid']
        counter = counter + 1

pheromones = np.ones((6, 6))

print(attractiveness)
print(pheromones)
print(waktu)
print(covid)

# Menampilkan nilai attractiveness serta pheromones
```

```python
def run_wpm(path):
    jaraks = np.array([])
    waktus = np.array([])
    covids = np.array([])
    for f in range(5): #Sesuai jumlah semut
        totalJarak = 0
        totalWaktu = 0
        totalCovid = 0
        for i in range(1, 6): #Calculate jarak, waktu, dan covid untuk setiap semut dari asal ke tujuan
            totalJarak = totalJarak + jarak[path[f][i-1]][path[f][i]]
            totalWaktu = totalWaktu + waktu[path[f][i-1]][path[f][i]]
            totalCovid = totalCovid + covid[path[f][i-1]][path[f][i]]
        if f == 0:
            jaraks = np.array([[totalJarak]])
            waktus = np.array([[totalWaktu]])
            covids = np.array([[totalCovid]])
        else:
            jaraks = np.append(jaraks, [[totalJarak]], axis=0)
            waktus = np.append(waktus, [[totalWaktu]], axis=0)
            covids = np.append(covids, [[totalCovid]], axis=0)
    path = np.append(path, jaraks, axis=1) #Kolom index  6
    path = np.append(path, waktus, axis=1) #Kolom index  7
    path = np.append(path, covids, axis=1) #Kolom index  8

    temp = path[:, 6].copy()
    
    #Masuk ke dalam WPM Utama

    #Tetapkan weight
    weight = np.array([3,1,2])
    weight = weight/np.sum(weight)
    
    path[:, 6] = np.amin(path[:, 6])/path[:, 6] #Jarak - Non Benefit
    path[:, 7] = np.amin(path[:, 7])/path[:, 7] #Waktu - Non benefit
    path[:, 8] = path[:, 8]/np.amax(path[:, 8]) #Covid - Benefit

    path[:, 6] = path[:, 6]**weight[0]
    path[:, 7] = path[:, 7]**weight[1]
    path[:, 8] = path[:, 8]**weight[2]

    sumPath = np.array([])
    for r in range(5):
        pro = 1
        for c in range(6, 9):
            pro = pro * path[r, c]
        sumPath = np.append(sumPath, [pro])
    path = np.append(path, np.transpose([sumPath]), axis=1)
    path = np.append(path, np.transpose([temp]), axis=1)
    return path
```

```python
def run_aco():
    for i in range(5): #Jumlah iterasi
        # print(f"i = {i}")
        final_path = np.array([])
        for j in range(5): #Jumlah semut/ambulan
            # print(f"j = {j}")
            chosen_path = np.array([0])
            for k in range(5):
                # print(f"k = {k}")
                accP = np.array([[],[]])

                for l in range(6):
                    # print(f"l = {l}")
                    isContinue = False
                    for temp in chosen_path: #Ngecek ud pernah dilewatin ato blm path ny
                        if temp == l:
                            isContinue = True
                            break
                    if isContinue == True: #kalo ud dilewatin, iterasi loop di skip
                        # print("Skip")
                        continue
                    p = (pheromones[chosen_path[-1]][l]**a) * (attractiveness[chosen_path[-1]][l]**b)
                    print(f"{chosen_path[-1]} - {l} p: {p}")
                    accP = np.append(accP, np.array([[p],[l]]), axis=1)
                total = 0
                for p in accP[0]:
                    total = total + p
                
                for n in range(accP[0].size):
                    accP[0][n] = accP[0][n] / total
                
                #print("\nSebelum akumulasi: ", accP[0], "/iteration")

                for n in reversed(range(accP[0].size)):
                    for m in range(n):
                        accP[0][n] = accP[0][n] + accP[0][m]
                    
                r = random.Random().random()

                #print("\nSesudah akumulasi: ", r, " - ", accP, "\n")

                for n in range(accP[0].size):
                    if r <= accP[0][n]:
                        chosen_path = np.append(chosen_path, [int(accP[1][n])])
                        print("Chosen Path: ", chosen_path)
                        break
                print()
            
            # print(chosen_path)
            # print(final_path)
            if j == 0:
                final_path = [chosen_path]
            else:
                final_path = np.append(final_path, [chosen_path], axis=0)
            
        wpm_result = run_wpm(final_path)

        print(wpm_result)

        #Update pheromones
        for r in range(6):
            for c in range(6):
                pheromones[r][c] = pheromones[r][c] - 0.05
                for ant in range(5):
                    for pointer in range(1, 6):
                        if wpm_result[ant][pointer] == c and wpm_result[ant][pointer-1] == r:
                            pheromones[r][c] = pheromones[r][c] - (1/wpm_result[ant][10])
        
        print(pheromones)
        if (pheromones < 0).any():
            finalIndex = np.argmax(wpm_result[:, 9])
            return wpm_result[finalIndex,[0,1,2,3,4,5]]

print("\nBest Path: ", run_aco())
    
```
